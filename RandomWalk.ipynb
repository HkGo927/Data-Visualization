```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Generate Step lenght for 100 steps
n_steps = 500

# Steps for x-axis
x_steps = np.random.normal(size = n_steps)

# Steps for y-axis
y_steps = np.random.normal(size = n_steps)
```

```python
# form the entire trajectory based on the defined steps
x = np.cumsum(x_steps)
y = np.cumsum(y_steps)
```

```python
# Draw the random walk
plt.figure(figsize=(6,6))
plt.plot(x, y, color='green', linewidth=2)
plt.grid(True)
plt.show()
```
