```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('salary.csv')
df.head()
```

```python
# Check distribution of YearsExperience
sns.kdeplot(data=df, x='YearsExperience')
plt.show()
```

```python
# Create a Jointplot to explore the relationship between experience and salary
sns.jointplot(data=df, x='YearsExperience', y='Salary', kind ='reg')
plt.show()
```

```python
# Using Scatterplot, display the relationship between x and y
sns.scatterplot(data=df, x='YearsExperience', y='Salary')
plt.show()
```

```python
# Model Building
x = df[['YearsExperience']]
y = df[['Salary']]
```

```python
# Split the data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
```

```python
# model prediction
y_pred =regr.predict(x_test)
y_pred
```

```python
y_test
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
r2_score(y_test, y_pred)
```

```python
# Visualize the model
sns.scatterplot(data=df, x='YearsExperience', y='Salary')
plt.plot(x_train, regr.predict(x_train), color='red')
plt.show()
```

```python
print('slope:', regr.coef_)
print('intercept:',regr.intercept_)
```
