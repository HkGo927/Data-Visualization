```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0,100,50)
c = x**2*2*x+6
print(x)
print(c)
```

```python
plt.figure(figsize=(8, 4))

# create a figure and axis
plt.plot(x, c, label='Plot of X^2+2*X+6', color='red')

# plot the data
plt.title('X^2+2*X+6')
plt.xlabel('x')
plt.ylabel('c')

plt.legend()
plt.grid(True)

plt.fill_between(x, c, color="red", alpha=0.2)

plt.show()
```
