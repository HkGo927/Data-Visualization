```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('energy.csv')
df.head()
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=df, x='Date', y='commercial', marker='o', markersize=12,
             linestyle='dashdot', mfc='#91e5f6', mec='grey')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
plt.figure(figsize=(15,7))

sns.lineplot(data=df, x='Date', y='commercial',
             marker ='p', markersize= 12, c='gold', mfc="yellow", mec="gold", label = "commercial")
sns.lineplot(data=df, x='Date', y='industrial',
             marker ='X', markersize= 12, c='grey', mfc="grey", mec="black", label = "industrial")

plt.gcf().autofmt_xdate()
plt.ylabel('$$$')
plt.legend()
plt.title('Commercial Energy Generations Over Time')
plt.grid(True)
plt.show()
```
