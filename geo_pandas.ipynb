```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt

from osgeo import gdal
gdal.SetConfigOption('SHAPE_RESTORE_SHX', 'YES')
```

```python
df1 = pd.read_csv('geo_data.csv')
df1
```

```python
gdf = geopandas.GeoDataFrame(df1, geometry=geopandas.points_from_xy(df1.Longitude, df1.Latitude))
gdf
```

```python
states = geopandas.read_file('usa-states-census-2014.shp')

fig, ax = plt.subplots(figsize=(14, 8))
states.plot(ax=ax, edgecolor='k', alpha=1, linewidth=1, cmap='Greens')
ax.set_facecolor('#a2d2ff')
plt.show()
```
