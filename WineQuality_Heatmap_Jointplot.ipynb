```python
# Task:
# 1. visualize the pair-wise coorelation coefficient
# 2. Extract all of the numeric variables
# 3. Create new data frame that EXCLUDE the quality
```

```python
new_df = df.iloc[:,0:11]
new_df.head()
```

```python
# Calculate the correlation Coefficient
coco = new_df.corr()
coco
```

```python
# Heatmap for coco

plt.figure(figsize=(10,10))
sns.heatmap(coco, square=True, cmap='RdBu_r', linewidth=0,5, vmin=-1, vmax=1)
plt.xticks(rotation=30)
plt.show()
```

```python
# Assume Quality is a numeric variable

df_coco = df.corr()
df_coco
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_coco, square=True, cmap='PiYG', linewidth=0.5, vmin=-1, vmax=1)
plt.xticks(rotation=30)
plt.show()
```

```python
# Mask
tri_matrix = np.triu(df_coco)
print(tri_matrix)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_coco, square=True, cmap='RdBu_r', linewidth=0.5, vmin=-1,vmax=1,
            mask=tri_matrix, annot=True, fmt=".1f")
plt.xticks(rotation=30)
plt.show()
```
