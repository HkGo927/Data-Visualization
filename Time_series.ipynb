```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('annual_gold.csv')
df.tail()
```

```python
# plot the Gold

plt.figure(figsize=(15,7))
plt.plot_date(x=df['Date'],y=df['Gold'],linestyle='--',linewidth=2,
              c='blue', marker='X', markersize=12, mfc='#ffbe0b', mec='grey')
plt.gcf().autofmt_xdate()
plt.title("Gold Price Change for the Past 30 Years")
plt.show()
```
