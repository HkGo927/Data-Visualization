```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

```python
df['Education'].value_counts()
```

```python
names = df['Education'].value_counts().index.tolist()
names
```

```python
values = df['Education'].value_counts().tolist()
```

```python
colors = ['#335c67', '#fff3b0', '#e09f3e', '#9e2a2b', '#540b0e']
```

```python
fig = plt.figure(figsize=(6,6))
plt.pie(values, labels=names, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
# This pie charts shows that Customers who has bacholer's degree has the largest.
```

```python
df['Kidhome'].value_counts()
```

```python
sns.histplot(data=df, x='Income', hue='Kidhome', kde=True, common_norm=True)
plt.show()
```

```python
sns.catplot(data=df, x='Kidhome', y='Income', kind='box')
plt.show()
```

```python
sns.scatterplot(data=df, x='NumStorePurchases', y='NumWebPurchases', hue='Kidhome')
plt.grid(True)
plt.show()
```

```python
# Meanwhile here is no Strong relationship between NumStorePurchases and WebPurchases, if you have One Kid at home,
#as NumStorePurchases increases, the NumWebPurchases slitly increase too.
# But there is no strong relationship between them.
```

```python
sns.catplot(data = df, x = 'Marital_Status', y = 'Income', kind = 'box', hue = 'Marital_Status', palette = 'Set1')
plt.xticks(rotation = 20)
plt.grid(True)
plt.show()
```

```python
sns.histplot(data=df, x='Income', hue='Marital_Status', kde=True,
             palette='Set3')
plt.show()
```

```python
sns.kdeplot(data = df, x = 'Income', hue = 'Marital_Status', fill = True, linewidth = 2, alpha = 0.2)
plt.show()
```

```python
sns.kdeplot(data = df, x = 'Income', hue = 'Marital_Status', common_norm = False, fill = True, linewidth = 2, alpha = 0.2)
plt.show()
```

```python
# This catplot graph shows a distribution of income across marital status.
# The result shows marital status "Alone" tends to show Lower Income Meanwhile "Married" and "Single" indicates the highest income.
```

```python
sns.catplot(data = df, x = 'Education', y = 'Income', kind = 'box', hue = 'Education', palette = 'Set1')
plt.xticks(rotation = 20)
plt.show()
```

```python
sns.histplot(data=df, x='Income', hue='Education', kde=True)
plt.show()
```

```python
# I selected a box plot visualization to find which education type of customers has the highest income.
# The result shows that PhD education tends to have the highest income with The Outliers. The master education type comes second. On the other hand,
# Basic education level shows the lowest income.
```
