```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
# Create a histogram for math_score
# kde: Kernal Density Estimate (smoother version of histplot)

sns.histplot(data=df, x='math_score', color='blue', kde=True)
plt.show()
```

```python
# What if I only need the curved line to display the distribution of math??

sns.kdeplot(data=df, x='math_score', fill=True, color='purple')
plt.show()
```

```python
# Visualize the math_score distribution across gender

sns.histplot(data=df, x='math_score', hue='gender',
             palette={'female':'lightpink', 'male':'skyblue'}, kde=True)
plt.show()
```

```python
# By kdeplot, Visualize the math_score distribution across gender.

sns.kdeplot(data=df, x='math_score', hue='gender',
            palette={'female':'pink', 'male':'green'}, fill=True,
            alpha=0.1)
plt.show()
```

```python
# the math_score distribution across ethnicity

sns.histplot(data=df, x='math_score', hue='ethnicity', kde=True,
             palette='Set3')
plt.show()
```

```python
# Using kdeplot(), visualize the math_score distribution across ethinicity
# common_norm=False: create curve based on ethnicity individually

sns.kdeplot(data=df, x='math_score', hue='ethnicity',
            palette='Set1', common_norm=False, fill=True, alpha=0.1,
            linewidth=2)
plt.show()
```

```python
# Create Boxplot for reading_score

sns.boxplot(data=df, x='reading_score', color='lightgreen')
plt.show()
```

```python
# Grouped boxplot
# visualize the reading_score across lunch by using boxplot

sns.catplot(data=df, x='lunch', y='reading_score', kind='box')
plt.grid(True)
plt.show()
```

```python
# Using Boxplot, visualize the reading_score across parental_education

sns.catplot(data=df, x='parental_education', y='reading_score', kind='box')
plt.xticks(rotation=20)
plt.show()
```

```python
# Using Scatter plot, find the relationship between two numeric variables. (math and reading score vs writing and reading score)

sns.scatterplot(data=df, x='math_score', y='reading_score', hue='gender',
                marker='x', s=200, palette={'female':'#f15bb5', 'male':'green'})
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x='writing_score', y='reading_score', hue='gender',
                marker='x', s=200, palette={'female':'#f15bb5', 'male':'blue'})
plt.grid(True)
plt.show()
```
