```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
numeric_df = df[['age', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']]
numeric_df.head()
```

```python
corr = numeric_df.corr()
corr
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(corr, square=True, cmap='bwr', vmin=-1, vmax=1, linewidth=0.5, annot=True)
plt.show()
```

```python
matrix = np.triu(corr)
matrix
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(corr, mask = matrix, square=True, cmap='bwr', vmin=-1, vmax=1, linewidth=0.5, annot=True)
```
